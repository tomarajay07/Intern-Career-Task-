# Are there specific categories with exceptional performance metrics

```python
threshold_suscribers=5000000      # Example: more than 5 million suscribers
threshold_visits=1000000          # Example: more than 1 million visits
threshold_likes=50000             # Example: more than 50,000 likes
threshold_comments=1000           # Example: more than 1,000 comments


# Filter categories based on exceptional performance metrics
exceptional_categories=dt[(dt['Suscribers']> threshold_suscribers)| (dt['Visits']> threshold_visits)| 
        (dt['Likes']> threshold_likes)| (dt['Comments']>threshold_comments)]

# Display the categories with exceptional performance
('Categories with Exceptional Performance')
(exceptional_categories['Categories'].unique())
```

# Brands and Collaborations:

```python
# Create a hypothetical metric 'BrandEngagementScore' by combining performance metrics
dt['BrandEngagementScore']= dt['Suscribers']+ 0.5 * dt['Visits']+0.2 * dt['Likes'] * dt['Comments']

# Set a threshold for high BrandEngagementScore
threshold_brand_engagement= 500000    # Example: more than 500,000 BrandEngagementScore

# Identify high-performance streamers based on BrandEngagementScore threshold
high_engagement_streamers = dt[dt['BrandEngagementScore'] > threshold_brand_engagement]

# Analyze the relationship between high performance and brand engagement
plt.figure(figsize=(10,6))
sns.scatterplot(x='Suscribers' , y='BrandEngagementScore' , data=high_engagement_streamers, hue='Visits', palette='viridis')
plt.title=('Relationship Between Suscribers, BrandEngagementScore, and Visits')
plt.xlabel=('Suscribers')
plt.ylabel=('BrandEngagementScore')
plt.show()
```

# Benchmarking:

Identify streamers with above-average performance in terms of subscribers, visits, likes, and comments

```python
# Calculate average metrics
average_suscribers = dt['Suscribers'].mean()
average_visits = dt['Visits'].mean()
average_likes = dt['Likes'].mean()
average_comments = dt['Comments'].mean()

# Identify streamers with above-average performance
above_average_streamers =dt[(dt['Suscribers']> average_suscribers) & (dt['Visits']>average_visits) & (dt['Likes']> 
    average_likes) & (dt['Comments']> average_comments)]

# Display information about above-average streamers
('Streamers with Above-Average Performance')
(above_average_streamers[['Username', 'Suscribers', 'Visits', 'Likes','Comments']])
```

Who are the top-performing content creators

```python
# Calculate the total engagement score for each content creator
dt['TotalEngagementScore']= dt['Suscribers'] + dt['Visits'] + dt['Likes'] + dt['Comments']

# Identify the top-performing content creators based on the total engagement score
top_performing_creators = dt.nlargest(10, 'TotalEngagementScore')

# Display information about the top-performing content creators
('Top-Performing Content Creators')
(top_performing_creators[['Username', 'TotalEngagementScore']])
```
